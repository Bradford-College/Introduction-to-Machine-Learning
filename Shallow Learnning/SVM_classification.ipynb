```python
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np    
```

```python
# Load the digits dataset from scikit-learn
digits = datasets.load_digits()
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.5, random_state=42)
```

```python
# Create a Support Vector Machine (SVM) classifier
svm_classifier = svm.SVC(gamma=0.001)
```

```python
# Train the classifier on the training data
svm_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = svm_classifier.predict(X_test)
```

```python
# Calculate and display the classification report
classification_report = metrics.classification_report(y_test, y_pred)
```

```python
print(classification_report)
```

```python
# Create a confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
print(confusion_matrix)
```

```python
# Visualize some sample predictions with their actual labels
images_and_predictions = list(zip(X_test, y_pred))
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for ax, (image, prediction) in zip(axes.flatten(), images_and_predictions[:10]):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title(f'Pred: {prediction}')
```
